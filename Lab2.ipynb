```python
cancer_b_data_filepath = './cancer_b.csv'
cancer_m_data_filepath = './cancer_m.csv'

cancer_b_data = pd.read_csv(cancer_b_data_filepath, index_col ='Id')
cancer_m_data = pd.read_csv(cancer_m_data_filepath, index_col ='Id')
```

```python
cancer_b_data.head()
```

En ambos conjuntos de datos cada fila corresponde a una imagen distinta.
 Cada conjunto tienen 31 columnas correspondientes a:

 1 columna "Diagnóstico" que clasifica a los tumores como begninos (aparecen como b o m )
 30 columnas que contienen diferentes medidad obenidad de las imágenes 

Investigaremos las diferencias
Crearemos 3 histogramas que muestren la distribución de los valores 'Area'

En una sola figura

```python
sns.displot(a=cancer_b_data['Area (mean)'], label='Tumor Benigno', kde= False)

sns.displot(a=cancer_m_data['Area (mean)'], label='Tumor Benigno', kde= False)

plt.title('Histograma de tumores benignos y malignos')
```

```python
sns.distplot(a=cancer_b_data['Area (mean)'], label='Tumor Benigno', kde=False)

sns.distplot(a=cancer_m_data['Area (mean)'], label='Tumor Maligno', kde=False)

plt.title('Histograma de tumores benignos y malignos')
plt.legend()
```

```python
plt.title('Gráfica KDE para tumores benignos y malignos')

sns.kdeplot(data=cancer_b_data['Radius (worst)'], shade=True, label='Benign Tumor')

sns.kdeplot(data=cancer_m_data['Radius (worst)'], shade=True, label='Malign Tumor')
```
